```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Load cleaned dataset
df = pd.read_csv("../data/all_f1_data.csv")
df.head()
```

```python
#loading in more information for predictive analysis 
df_races   = pd.read_csv("../data/races.csv")
df_qual    = pd.read_csv("../data/qualifying.csv")
df_results = pd.read_csv("../data/results.csv")
df_drivers = pd.read_csv("../data/drivers.csv")
df_base    = pd.read_csv("../data/all_f1_data.csv")
```

```python
# Filter races 2014-2024
df_races = df_races[df_races['year'].between(2014, 2024)]
race_ids = df_races['raceId'].unique()
df_qual = df_qual[df_qual['raceId'].isin(race_ids)]
df_results = df_results[df_results['raceId'].isin(race_ids)]
```

```python
# Merge driver names
df_qual = df_qual.merge(df_drivers[['driverId','forename','surname']], on='driverId', how='left')
df_qual['driverName'] = df_qual['forename'] + " " + df_qual['surname']
df_results = df_results.merge(df_drivers[['driverId','forename','surname']], on='driverId', how='left')
df_results['driverName'] = df_results['forename'] + " " + df_results['surname']
```

```python
# Merge qualifying and results for each driver/race
df_merge = df_results.merge(
    df_qual[['raceId','driverId','position']], 
    on=['raceId','driverId'], 
    how='left',
    suffixes=('_finish','_qualify')
)
```

```python
# Rename columns for clarity
df_merge.rename(columns={'position_qualify':'gridPosition','position_finish':'finishPosition'}, inplace=True)
```

```python
# Add year and Grand Prix name
df_merge = df_merge.merge(df_races[['raceId','year','name']], on='raceId', how='left')
df_merge.rename(columns={'name':'Grand Prix'}, inplace=True)
df_merge.to_csv("../data/merged_f1_data.csv", index=False)
df_merge.head()
```
