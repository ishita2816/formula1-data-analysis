```python
#importing data for 2014
# Load the 2014 race results CSV
results_2014 = pd.read_csv("../data/f1_2014(Sheet1).csv")
# Preview the first few rows
results_2014.head()
```

```python
# Combine all yearly DataFrames into one
all_data = pd.concat(
    [results_2014, results_2015, results_2016, results_2017,
     results_2018, results_2019, results_2020, results_2021,
     results_2022, results_2023, results_2024],
    ignore_index=True
)

# Verify combined data
print("Total races combined:", all_data.shape[0])
all_data.head()
```

```python
# Count wins per driver
driver_wins = all_data["WINNER"].value_counts().head(10)
print(driver_wins)
```

```python
# plotting wins of each driver 
driver_wins.plot(kind='bar', figsize=(8,4), color='skyblue', title='Top 10 Drivers by Wins (2014–2024)')
plt.xlabel("Driver")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45)
plt.show()
plt.savefig("../visuals/top_drivers.png", bbox_inches='tight')
```

```python
# Count wins per team
team_wins = all_data["TEAM"].value_counts().head(10)
print(team_wins)
```

```python
#plotting wins of each team
team_wins.plot(kind='bar', figsize=(8,4), color='pink', title='Top 10 Teams by Wins (2014–2024)')
plt.xlabel("Team")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45)
plt.show()
plt.savefig("../visuals/top_teams.png", bbox_inches='tight')
```

```python
# adding year coulmn so we can analyze data year wise 
all_data = pd.concat(
    [
        results_2014.assign(YEAR=2014),
        results_2015.assign(YEAR=2015),
        results_2016.assign(YEAR=2016),
        results_2017.assign(YEAR=2017),
        results_2018.assign(YEAR=2018),
        results_2019.assign(YEAR=2019),
        results_2020.assign(YEAR=2020),
        results_2021.assign(YEAR=2021),
        results_2022.assign(YEAR=2022),
        results_2023.assign(YEAR=2023),
        results_2024.assign(YEAR=2024)
    ],
    ignore_index=True
)
```

```python
# Count races per year
races_per_year = all_data["YEAR"].value_counts().head(10)
print(races_per_year)
```

```python
#plotting races per year 
races_per_year.plot(kind='line', marker='o', figsize=(8,4), title='Number of Races per Season')
plt.xlabel("Year")
plt.ylabel("Number of Races")
plt.grid(True)
plt.show()
plt.savefig("../visuals/races_per_season.png", bbox_inches='tight')
```
