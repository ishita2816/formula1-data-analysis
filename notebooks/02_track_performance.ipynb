# Track Performance Analysis (2014–2024)

This notebook analyzes which drivers and teams win at which tracks (Grand Prix) from 2014–2024. 
We use heatmaps to visualize patterns of dominance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading combined data 
all_data = pd.read_csv("../data/all_f1_data.csv")
all_data.head()
```

```python
track_driver_wins = all_data.groupby(['GRAND PRIX', 'WINNER']).size().unstack(fill_value=0)
print(track_driver_wins)
plt.figure(figsize=(12,8))
sns.heatmap(track_driver_wins, cmap="YlGnBu", linewidths=0.5)
plt.title("Wins per Track by Driver (2014–2024)")
plt.xlabel("Driver")
plt.ylabel("Grand Prix")
plt.tight_layout()
plt.savefig("../visuals/wins_per_track_driver.png", bbox_inches='tight')
plt.show()
```

```python
track_team_wins = all_data.groupby(['GRAND PRIX', 'TEAM']).size().unstack(fill_value=0)
print(track_team_wins)
plt.figure(figsize=(12,8))
sns.heatmap(track_team_wins, cmap="magma", linewidths=0.5)
plt.title("Wins per Track by Teams (2014–2024)")
plt.xlabel("Teams")
plt.ylabel("Grand Prix")
plt.tight_layout()
plt.savefig("../visuals/wins_per_team_driver.png", bbox_inches='tight')
plt.show()
```

## analyzing results so far
### track × driver heatmap data
Looking at driver performance across different tracks, we can see clear patterns of dominance. Lewis Hamilton consistently wins at multiple circuits, including Abu Dhabi, Great Britain, and Spain, reinforcing his overall season dominance. Max Verstappen performs strongly at tracks like Austria, Mexico, and Spain, reflecting his rise in recent seasons. Other drivers like Charles Leclerc, Carlos Sainz, and Daniel Ricciardo have more sporadic wins, often limited to one or two circuits, indicating targeted strengths or team-dependent performance. Overall, a small group of top drivers dominates most races, while others occasionally secure victories at specific tracks, highlighting both driver skill and team strategy in Formula 1.

### track x team heatmap data 
Looking at wins by team across different Grand Prix, Mercedes clearly dominates most circuits, consistently winning at Abu Dhabi, Great Britain, Japan, and Spain. Red Bull Racing, across its various engine partnerships, shows strong performances at tracks like Austria, Mexico, and Monaco, reflecting periods of competitiveness in recent seasons. Ferrari has notable successes at Australia, Bahrain, and Italy, while teams like McLaren, AlphaTauri, and Racing Point achieve occasional wins. Overall, a few top teams capture the majority of race victories at nearly every track, highlighting how team performance and strategy are critical in shaping the outcomes of Formula 1 races.
